**Step 3 : Model Implementation**

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, recall_score, accuracy_score, precision_score, confusion_matrix, f1_score, roc_auc_score
```

```python
x = heart.iloc[:,:-1]
y = heart.iloc[:,-1]

```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.20,random_state=40)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

### *KNN Classifier*

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train, y_train)
y_train_predict_knn = knn.predict(x_train)
y_test_predict_knn = knn.predict(x_test)
```

*Naive Bayes Classifier (Gaussian Naive Bayes)*

```python

nbc = GaussianNB()
nbc.fit(x_train, y_train)
y_train_predict_nbc = nbc.predict(x_train)
y_test_predict_nbc = nbc.predict(x_test)
```

# *Decision Tree Classifier*

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train_new = scaler.fit_transform(x_train)
x_test_new = scaler.transform(x_test)
```

```python

dtc = DecisionTreeClassifier(random_state=40, max_depth=6)
dtc.fit(x_train_new, y_train)
y_train_predict_dtc = dtc.predict(x_train)
y_test_predict_dtc = dtc.predict(x_test_new)
```

# Model Evaluation

```python
# Classification metrics for KNN Classifier
print("KNN Classifier:")
print("Train Accuracy:", accuracy_score(y_train, y_train_predict_knn))
print("Test Accuracy:", accuracy_score(y_test, y_test_predict_knn))
print("Recall:", recall_score(y_test, y_test_predict_knn))
print("Precision:", precision_score(y_test, y_test_predict_knn))
print("F1 Score:", f1_score(y_test, y_test_predict_knn))
print("ROC AUC Score:", roc_auc_score(y_test, y_test_predict_knn))
print(confusion_matrix(y_test, y_test_predict_knn))
print(classification_report(y_test, y_test_predict_knn))
```

```python
# Classification metrics for DTC  Classifier
print("dtc Classifier:")
print("Train Accuracy:", accuracy_score(y_train, y_train_predict_dtc))
print("Test Accuracy:", accuracy_score(y_test, y_test_predict_dtc))
print("Recall:", recall_score(y_test, y_test_predict_dtc))
print("Precision:", precision_score(y_test, y_test_predict_dtc))
print("F1 Score:", f1_score(y_test, y_test_predict_dtc))
print("ROC AUC Score:", roc_auc_score(y_test, y_test_predict_dtc))
print(confusion_matrix(y_test, y_test_predict_dtc))
print(classification_report(y_test, y_test_predict_dtc))
```

```python
#Gaussian Naive Bayes Classifier Model Performance:
print("nbc Performance Matrix :")
print("Train Accuracy:", accuracy_score(y_train, y_train_predict_nbc))
print("Test Accuracy:", accuracy_score(y_test, y_test_predict_nbc))
print("Recall:", recall_score(y_test, y_test_predict_nbc))
print("Precision:", precision_score(y_test, y_test_predict_nbc))
print("F1 Score:", f1_score(y_test, y_test_predict_nbc))
print("ROC AUC Score:", roc_auc_score(y_test, y_test_predict_nbc))
print(confusion_matrix(y_test, y_test_predict_nbc))
print(classification_report(y_test, y_test_predict_nbc))
```

```python
from sklearn.model_selection import cross_val_score

# Define the models
models = {
    "KNN": KNeighborsClassifier(n_neighbors=5),
    "Naive Bayes": GaussianNB(),
    "Decision Tree": DecisionTreeClassifier(random_state=40, max_depth=6)
}

# Loop through the models
for model_name, model in models.items():
    print(f"Model: {model_name}")

    # Cross-validation scores for training set
    cv_scores = cross_val_score(model, x_train, y_train, cv=5, scoring="accuracy")
    print("Cross-Validation Scores (Training):", cv_scores)
    print("Mean CV Score (Training):", cv_scores.mean())

    # Train the model
    model.fit(x_train, y_train)

    # Training accuracy
    train_accuracy = accuracy_score(y_train, model.predict(x_train))
    print("Training Accuracy:", train_accuracy)

    # Testing accuracy
    test_accuracy = accuracy_score(y_test, model.predict(x_test))
    print("Testing Accuracy:", test_accuracy)

    # Check for overfitting
    if train_accuracy > test_accuracy:
        print("Model is Overfitted")
    else:
        print("Model is Not Overfitted")

    print("-" * 50)
```

```python
from sklearn.model_selection import cross_val_score

# Define the models
models = {
    "KNN": KNeighborsClassifier(n_neighbors=5),
    "Naive Bayes": GaussianNB(),
    "Decision Tree": DecisionTreeClassifier(random_state=40, max_depth=6)
}

# Loop through the models
for model_name, model in models.items():
    print(f"Model: {model_name}")

    # Cross-validation scores for both training and testing sets
    cv_train_scores = cross_val_score(model, x_train, y_train, cv=10, scoring="accuracy")
    cv_test_scores = cross_val_score(model, x_test, y_test, cv=10, scoring="accuracy")

    # Print train and test accuracies for each fold
    for fold in range(10):
        print(f"Fold {fold+1}: Train Accuracy = {cv_train_scores[fold]}, Test Accuracy = {cv_test_scores[fold]}")

    # Calculate and print average accuracy
    avg_train_accuracy = cv_train_scores.mean()
    avg_test_accuracy = cv_test_scores.mean()
    print(f"Avg Train Accuracy = {avg_train_accuracy}, Avg Test Accuracy = {avg_test_accuracy}")

    print("-" * 50)
```
